```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
from time import sleep
```

```python
deriver=webdriver.Chrome()
deriver.get('https://iranfile.ir/properties/buy-apartment?page=357')
sleep(5)
html_cor=deriver.page_source
data_cor=pd.read_html(html_cor)
data_fram=data_cor[0]
for i in range(900):
    print(i,'/900')
    try:
        deriver.find_element(By.CLASS_NAME,"next ").click()
    except:
        sleep(10)
        try:
            deriver.find_element(By.CLASS_NAME,"next ").click()
        except:
            continue
    sleep(5)
    try:
        html_secandery=pd.read_html(deriver.page_source)
    except:
        try:
            sleep(10)
            html_secandery=pd.read_html(deriver.page_source)
        except:
            continue       

    data_fram_secendery=html_secandery[0]
    data_fram=pd.concat([data_fram,data_fram_secendery])    
    
```

```python
index_data_fram=[*range(len(data_fram))]
data_fram['number']=index_data_fram
data_fram.set_index(data_fram['number'], inplace=True)
```
